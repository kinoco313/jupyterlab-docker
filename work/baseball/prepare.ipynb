```python
# 12球団の打撃成績を10年分（2014年から2023年）集める
# ライブラリの読み込み

import pandas as pd
```

```python
# 2023パリーグ
url_23p = 'https://npb.jp/bis/2023/stats/tmb_p.html'
df_23p = pd.read_html(url_23p)[0]

# 年度の列を追加（後で他の年度と結合するので）
df_23p['年度'] = 2023

# 確認
df_23p
```

```python
# 2023セリーグ
url_23c = 'https://npb.jp/bis/2023/stats/tmb_c.html'
df_23c = pd.read_html(url_23c)[0]
df_23c['年度'] = 2023

# 確認
df_23c
```

```python
# とりあえず2023のセパを結合してみる
df_23 = pd.concat([df_23p, df_23c], axis=0, ignore_index=True)
df_23
```
